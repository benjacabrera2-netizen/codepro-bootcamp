```python
#Challenge 16

#

letra01 = input('Introduzca letra 3')
letra02 = input('Introduzca letra 4')

def comparador_de_letras(word1, word2):
    letra03 = word1[0]
    letra04 = word2[0]
    
    while True:
        if letra03[0] == letra04[0]:
            print(f'Las letras {letra03[0]} y {letra04[0]} si son letras iguales.')
            break
        else:
           print(f'Las letras {letra03[0]} y {letra04[0]} no son letras iguales.')
           letra03 = input('otra vez')
           letra04 = input('otra vez')

comparador_de_letras(letra01, letra02)

```

```python
# Challenge 18 游뱤 - 20 min

# Crear funci칩n que revise si una palabra contiene una letra. En caso de que la contenga, retornar la letra entre corchetes [].
# De lo contrario, retornar solo la letra.

la_letra = 'b'
user_input= input('palabrinha')

def the_word(word1):
    for i17 in word1:
        if i17 == la_letra:
            return [i17]
    return la_letra
        
the_word(user_input)
```
