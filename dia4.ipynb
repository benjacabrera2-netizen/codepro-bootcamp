```python
# Challenge ðŸ¤º - 15 min

# Crear dos mÃ©todos para nuestro animal y ejecutar, debe ser algo especÃ­fico de ese animal.
    # Ejemplo: "Cuando hablo, digo X".
    # Ejemplo: "Cuando me muevo, hago X".

class Leon:
    especie = 'felino'

    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color
    
    def calcula(self, metros):
        calculo0 = metros / 2
        print(f'Cuando me dan un numero, les digo la mitad, en este caso: {calculo0}.')
    
    def dormir(self, horas):
        print(f'Cuando duermo, duermo por {horas} horas.')

leon1 = Leon('Mufasa', 'Amarillo')
print(leon1.nombre + ' ' + leon1.color)
leon1.calcula(10)
leon1.dormir(24)
```
