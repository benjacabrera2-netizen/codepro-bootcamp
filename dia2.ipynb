```python
# Challenge 8 游뱤 - 15m

# Escribe un programa que pregunte al usuario por su edad y determine si tiene edad suficiente para comprar y consumir alcohol.
# Si el usuario tiene mas de 18 pero menos de 21 impirmir "Podes tomar pero no comprar"
# Si el usuario tiene menos de 18 imprimir "Juguito nomas"
# Si el usuario tiene mas de 21 imprimir "Hace lo que quieras"

#Obs: se quieren validar dos cosas al mismo tiempo.

edad_usuario = int(input('Ingrese su edad.'))

if edad_usuario > 17 and edad_usuario < 21:
    print("Podes tomar pero no comprar.")
elif edad_usuario > 20:
    print("Hace lo que quieras.")
elif edad_usuario < 18:
    print("Juguito nomas.")
else:
    print("Probablemente ya este ebrio.")
```

```python
#Challenge 9 游뱤 - 20m

#Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:
# Entre -10 y 18 grados, que imprima "Fresquete".
# Entre 19 y 30 grados, que imprima "Calor'i".
# Entre 31 y 45 grados, que imprima "Hakuuuuuu".
# Para todo lo dem치s, imprimir "칌amano mba'e!".

import math
temperatura = math.floor(float(input('Temperatura.')))
print(f'La temperatura actual es: {temperatura} Celsius.\nSe declara oficialmente:')

if temperatura > -11 and temperatura < 19:
    print("Fresquete")
elif temperatura > 18 and temperatura < 31:
    print("Calor'i")
elif temperatura > 30 and temperatura < 46:
    print("Hakuuuuuu")
else:
    print("칌amano mba'e!")
```
